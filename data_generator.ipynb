```python
import numpy as np
import matplotlib.pyplot as plt
import pickle
```

```python
def generate_sine_series(num_points, frequency, amplitude,
                         noise_factor=0, volume_factor=1000,
                         plot=False):
    
    x = np.linspace(0, 100 * np.pi, num_points)

    sine_wave = amplitude * (1.1+np.sin(frequency * x))
    sine_wave2 = amplitude/2 * (np.sin(frequency*3 * x))
    abs_sine_wave = amplitude/4 * (np.abs(np.sin(frequency*4) * x))

    noise = noise_factor * (amplitude/4) * np.random.randn(num_points)
    sine_series = sine_wave + sine_wave2 + abs_sine_wave + noise

    random_volume = volume_factor * (np.abs(1.1+np.random.randn(num_points)))

    if plot:
        plt.plot(x[100:200], sine_series[100:200])
        plt.title(f'Sine Wave (Frequency={frequency}, Amplitude={amplitude})')
        plt.show()

        plt.plot(x[100:200], random_volume[100:200])
        plt.title(f'Volume generated (with volume factor {volume_factor})')
        plt.show()
    return np.column_stack([sine_series, random_volume])

sine_series = generate_sine_series(num_points=int(1e4), frequency=3,
                                    amplitude=100, noise_factor=0.4,
                                    volume_factor = 1000, plot=True)
sine_series.shape
```

```python
with open("data/sin/sin.pickle", 'wb') as f:
    pickle.dump(sine_series, f)
```
